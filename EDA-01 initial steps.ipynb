```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
numeric_features
```

```python
categorical_features
```

```python
# categorical columns
plt.figure(figsize=(20,30))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=50, fontweight='bold', alpha=1, y=1.025)
category = [' State-gov',' Bachelors',' Never-married',' Adm-clerical',' Not-in-family',' White',' Male',' United-States',' <=50K']
for i in range(0, len(category)):
    plt.subplot(7,2, i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=90)
    plt.tight_layout()
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=18, fontweight='bold', alpha=0.5, y=1.025)

for i in range(0, len(numeric_features)):
    plt.subplot(3, 2, i+1)
    sns.kdeplot(data=df[numeric_features[i]],shade=True, color='r',bw=1.5)
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```
